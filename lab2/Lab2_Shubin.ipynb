```python
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
angles = [-20, -60, -10, 30, 60, 70, -10, -30, -120, -100, -80, 20, 40, -60, -10, 20, 30, -80, 60, 70]
```

```python
ecdf = ECDF(angles)
ecdf.x[0] = -120
plt.hlines(0, -200, -120, 'blue')
plt.hlines(1, 70, 200, 'blue')
plt.plot(ecdf.x, ecdf.y)
plt.axis([-125, 75, -0.1, 1.1])
plt.grid()
plt.show()
```

```python
sns.histplot(angles, binwidth=10)
```
